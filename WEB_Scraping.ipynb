```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://myanimelist.net/topanime.php'  ## top_anime_url = url + '/topanime.php'
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
headers = soup.find('tr', class_ = 'table-header')
headers.find_all('td')
```

```python
headers = soup.find_all('tr', class_ = 'ranking-list')
headers
```

```python
# headers[0].find('td', class_ = "rank ac").find('span').text
# headers[0].find('div', class_="di-ib clearfix").find('a').text
# headers[0].find('td', class_="score ac fs14").find('span').text
# var = headers[5].find('div', class_ = "information di-ib mt4").text.strip().split('\n')
# var
```

```python
top_anime = []
for row in headers:
    ranking = [
         row.find('td', class_ = "rank ac").find('span').text,
         row.find('div', class_="di-ib clearfix").find('a').text,
         row.find('td', class_="score ac fs14").find('span').text,
    ]
    top_anime.append(ranking)
```

```python
release = soup.find_all('div', class_ = 'information di-ib mt4')

mytable_title =[title.text.strip().split('\n') for title in release]
mytable_title
cleaned_data_strip = [[s.strip() for s in sublist] for sublist in mytable_title]
cleaned_data_strip[:2]
```

```python
joined_list = [a + b for a, b in zip(top_anime, cleaned_data_strip)]
joined_list[:2]
```

```python
df = pd.DataFrame(joined_list,columns =['Rank','Anime ','Rating','Episodes','Dates','Members'])
df.set_index('Rank', inplace=True)
df
```
